```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

pd.set_option('display.max_rows',300)
pd.set_option('display.max_column',20)
```

```python
dataset = pd.read_csv("data/dataset_with_features.csv")
print("dataset shape",dataset.shape)
dataset.head(5)
```

```python
X_origin = dataset.iloc[:,[3,4] + list(range(8,dataset.shape[1]))]
#filling NaN
# median_Concen = X_origin['Concentration(mg/ml)'].median()
mean_concen= X_origin['Conc(mg/ml)'].mean()
X_origin.iloc[:,1:2] = X_origin['Conc(mg/ml)'].fillna(mean_concen)
print("X_origin shape:",X_origin.shape)
X_origin.head()
```

```python
#initial pearson heatmap
corr = X_origin.corr(method='pearson')
mask = np.zeros_like(corr)
plt.rcParams['font.family']="Arial"
mask[np.triu_indices_from(mask)] = True
with sns.axes_style("white"):
    f, ax = plt.subplots(figsize=(16,12))
    ax = sns.heatmap(corr, annot=True,annot_kws={"size": 10},fmt=".2f",
                     cmap='RdBu_r',linewidths=1,vmin=-1,vmax=1,mask=mask, square=True)
plt.title(" ")
plt.savefig("output_figures/Initial_peason_corr.png",dpi = 300)
plt.show() 
```

```python
#remove hight correlation features : Heavy_atoms/Valence_electrons/Num_Carbon
# X_origin_proc = X_origin.drop(['Heavy_atoms','Val_electrons','Topo'],axis=1)
X_origin_proc = X_origin.drop(['Heavy_atoms','Carbon','Val_electrons','Topo'],axis=1)

corr = X_origin_proc.corr(method='pearson')
mask = np.zeros_like(corr)
plt.rcParams['font.family']="Arial"
mask[np.triu_indices_from(mask)] = True
with sns.axes_style("white"):
    f, ax = plt.subplots(figsize=(16,12))
    ax = sns.heatmap(corr, annot=True,annot_kws={"size": 10},fmt=".2f",
                     cmap='RdBu_r',linewidths=1,vmin=-1,vmax=1,mask=mask, square=True)
plt.title(" ")
plt.savefig("output_figures/pocessed_peason_corr.png",dpi = 300)
plt.show()  
```

```python
# y_origin = df['EQE enhanced ratio']
y_origin = dataset['EQE enhanced ratio']
y_origin = pd.DataFrame(y_origin)
y_origin.shape
```

```python
fig, ax = plt.subplots()
width_size = (np.max(y_origin) - np.min(y_origin))/30
ax.hist(y_origin, width = width_size, bins = 20, color= 'deepskyblue')
locator = ticker.MultipleLocator(base=0.5)  # setting scale of x = 0.1
ax.xaxis.set_major_locator(locator)
plt.ylabel('Counts of EQE', fontsize = 10)
plt.xlabel('target(enhanced ratio)', fontsize = 10)
# plt.xlim(0,3.0)
plt.savefig("output_figures/target_dis.png",dpi = 300)
plt.show()
```

```python
#acculating statistics
n_bins = 5
y_bins = pd.qcut(y_origin.values.ravel(), n_bins)

quantile_counts = y_bins.value_counts()

plt.figure(figsize=(6, 4))
quantile_counts.plot(kind='bar', color='deepskyblue',alpha = 0.8)
plt.title('Histogram of Data by Quantiles')
plt.xlabel('Quantile Bins',fontsize = 10)
plt.ylabel('Frequency',fontsize = 10)
plt.xticks(rotation=30)
plt.show()
```
